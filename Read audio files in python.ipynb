```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from glob import glob

import librosa as lr
```

```python
# Set directory for source files

data_dir = './whitesound'
audio_files = glob(data_dir + '/*.wav')

len(audio_files)
```

```python
# Read in the first audio file, create the time array(timeline)
audio, sfreq = lr.load(audio_files[3])
time = np.arange(0, len(audio)) / sfreq

time
```

```python
# Plot audio over time
fig, ax = plt.subplots()
ax.plot(time, audio)
ax.set(xlabel='Time(s)', ylabel='Sound Amplitude')
plt.show()
```

```python
%%time

# Perform the reading through all the audio files
for file in range(0, len(audio_files), 1):
    # Read in the first audio file, create the time array(timeline)
    audio, sfreq = lr.load(audio_files[file])
    time = np.arange(0, len(audio)) / sfreq
    
    # Plot audio over time
    fig, ax = plt.subplots()
    ax.plot(time, audio)
    ax.set(xlabel='Time(s)', ylabel='Sound Amplitude')
    plt.show()
```
